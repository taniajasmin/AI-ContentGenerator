```python
from gtts import gTTS
from IPython.display import Audio
```

```python
speech_text = input("Enter the text you want your avatar to say: ")

# Convert text to speech
tts = gTTS(text=speech_text, lang='en')
audio_file_path = "avatar_speech.mp3"
tts.save(audio_file_path)

# Play the audio in Colab
print("Playing the audio:")
display(Audio(audio_file_path, autoplay=True))

# # Automatically download the audio file
# print("Downloading the audio file...")
# files.download(audio_file_path)
```
